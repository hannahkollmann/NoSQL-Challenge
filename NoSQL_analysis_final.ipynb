### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query_1 = {'scores.Hygiene' : 20}


# Use count_documents to display the number of documents in the result
count_1 = establishments.count_documents(query_1)
print(f"There are {count_1} restaurants with a hygiene score of 20.")


# Display the first document in the results using pprint
pprint(establishments.find_one(query_1))
```

```python
# Convert the result to a Pandas DataFrame
query_1_df = pd.DataFrame(establishments.find(query_1))


# Display the number of rows in the DataFrame
print(f"There are {len(query_1_df)} rows this dataframe.")


# Display the first 10 rows of the DataFrame
query_1_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_2 = {'LocalAuthorityName':{'$regex':'London'},'RatingValue': {'$gte':4}}


# Use count_documents to display the number of documents in the result
count_2 = establishments.count_documents(query_2)
print(f"There are {count_2} establishments in London that have a RatingValue greater than or equal to 4.")


# Display the first document in the results using pprint
pprint(establishments.find_one(query_2))
```

```python
# Convert the result to a Pandas DataFrame
query_2_df = pd.DataFrame(establishments.find(query_2))

# Display the number of rows in the DataFrame
print(f"There are {len(query_2_df)} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
query_2_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# What is the lat and long of Penong Flavours?
print(establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.latitude', 'geocode.longitude'}))
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

query_3 = {'RatingValue': 5, 
         'geocode.latitude': {'$lte': (latitude + degree_search), '$gte': (latitude - degree_search)}, 
         'geocode.longitude': {'$lte': (longitude + degree_search), '$gte': (longitude - degree_search)} 
         }
sort =  [('scores.Hygiene', 1)]

limit = 5

# Print the results
for result in establishments.find(query_3).sort(sort).limit(5):
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
pd.DataFrame(establishments.find(query_3).sort(sort).limit(5))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match': {'scores.Hygiene': 0}}


group_query = {'$group': {'_id': '$LocalAuthorityName', 
                          'count': {'$sum': 1}
                         }
              }

sort_query = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_query]

results = list(establishments.aggregate(pipeline))


# Print the number of documents in the result
print(f"There are {len(results)} documents in the result.")


# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
query_4_df = pd.json_normalize(results)


# Display the number of rows in the DataFrame
print(f"There are {len(query_4_df)} rows in the DataFrame.")


# Display the first 10 rows of the DataFrame
query_4_df.head(10)
```
